```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Make some random data
X, y = make_blobs(n_samples=300, cluster_std=4.00, random_state=42)
```

```python
X
```

```python
X[:, 1]
```

```python
y
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

### Elbow Method

```python
from sklearn.cluster import KMeans
wcss = [KMeans(n_clusters=i).fit(X).inertia_ for i in range(1, 11)]

wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
k_means = KMeans(n_clusters=3)
pred_y = k_means.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=30, c='red')
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=k_means.labels_.astype(float))
plt.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s=30, c='red')
plt.show()
```
