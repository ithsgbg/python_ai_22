Data set is from [UCI](https://archive.ics.uci.edu/ml/datasets/Student+Performance)

```python
import pandas as pd

df = pd.read_csv('./data/student-mat.csv', sep=';')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.G3.min()
```

```python
sex = pd.get_dummies(df.sex, prefix='sex')
df.drop('sex', axis=1, inplace=True)
df = df.join(sex)
```

```python
df.head(10)
```

```python
df.corr()
```

```python
df_small = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]
df_small.head()
```

```python
df_small.corr()
```

```python
import seaborn as sns

sns.heatmap(df_small.corr(), annot=True, fmt='.2f')
```

```python
import numpy as np

X = np.array(df_small.drop(["G3"], axis=1))
y = np.array(df['G3'])
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=34)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
linear_model.score(x_test, y_test)
```

```python
linear_model.coef_
```

```python
linear_model.intercept_
```

```python
predictions = linear_model.predict(x_test)
predictions
```

```python
for i in range(len(predictions)):
    print(f'Pred: {predictions[i]} - y: {y_test[i]}')
```
